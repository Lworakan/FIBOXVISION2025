```python
import pandas as pd

df = pd.read_csv('./output/merged_PLC_tracking.csv')
df
```

```python
# rename columns  Distance_RealSense
```

```python
# new dataframe get only depth and accealdatax and ground_truth

# Timestamp	Frame	X_min	Y_min	X_max	Y_max	Confidence	Average_Depth_m	area	ground_truth
# confidence,bbox_x1,bbox_y1,bbox_x2,bbox_y2,bbox_center_x,bbox_center_y,bbox_width,bbox_height,rel_x,rel_y,mean_depth,median_depth,min_depth,max_depth,std_depth,valid_pixels,total_pixels,fill_ratio,center_depth_raw,center_depth_meters,ground_truth	
data = df[['confidence', 'bbox_x1', 'bbox_y1', 'bbox_x2', 'bbox_y2',
'bbox_center_x', 'bbox_center_y', 'bbox_width', 'bbox_height',
           'rel_x', 'rel_y', 'mean_depth', 'median_depth', 'min_depth',
           'max_depth', 'std_depth', 'valid_pixels', 'total_pixels',
           'fill_ratio', 'center_depth_raw', 'center_depth_meters',
           'ground_truth']]
data
```

```python
# data[Average_Depth_m] = data['Distance_RealSense']
data = data.rename(columns={'Distance_RealSense': 'center_depth_meters'})
```

```python
from pycaret.regression import *
# s = setup(df, target = 'ground_truth', session_id = 123, use_gpu=True, polynomial_features=True,polynomial_degree=2)
s = setup(data, target = 'ground_truth', session_id = 123, use_gpu=True, polynomial_features=True)
```

```python
best = compare_models()

```

```python
# functional API
evaluate_model(best)
```

```python
plot_model(best, plot = 'residuals')
```
