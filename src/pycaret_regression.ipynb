```python
import pandas as pd

df = pd.read_csv('../Callback/merged_output.csv')
df
```

```python
# new dataframe get only depth and accealdatax and ground_truth
data = df[['depth', 'ground_truth']]
data
```

```python
from pycaret.regression import *
s = setup(data, target = 'ground_truth', session_id = 123, use_gpu=True, polynomial_features=True,polynomial_degree=2)
```

```python
best = compare_models()

```

```python
# functional API
evaluate_model(best)
```

```python
plot_model(best, plot = 'residuals')
```

```python
plot_model(best, plot = 'feature')
```

```python
predict_model(best)
```

```python
predictions = predict_model(best, data=data)
```
